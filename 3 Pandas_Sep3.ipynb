#### concate and merge

```python
df.drop(['gender', 'cgpa'], axis = 1, inplace = True)
```

```python
df
```

```python
# create a newdf roll, gender, cgpa columns
# roll column is same as df
```

```python
newdf = pd.DataFrame({
    'roll': range(101, 111),
    'gender': gender,
    'cgpa': (df['marks']/10) + 1.5
})
```

```python
newdf
```

```python
# merge

pd.merge(df, newdf, on = 'roll')
```
